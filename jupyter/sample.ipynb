# Short cut command

---
## Command Mode
`m`
markdown mode

`j k`
move cell 'down' or 'up'

`a b`
create cell 'abobe' or 'below'

`dd`
delete cell

`l`
view line

---
## Insert Mode
`Esc`
exit insert mode

---
## Command and Insert Mode
`Shift + Enter`
execute cell
